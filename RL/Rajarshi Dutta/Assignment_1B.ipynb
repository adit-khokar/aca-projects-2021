```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#first graph
x=np.arange(0,100,10)
y= np.sin(x**2)+ np.log(x)+ np.cos(x)
plt.figure(figsize=(10,6))
plt.title('First Graph')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.plot(x,y, 'b^--',label=['x','y'], color='black')
plt.legend()
plt.grid()
```

```python
#second graph->pie plot
prices=[250,450,220,656,100,111,778,150]
plt.figure(figsize=(10,6))
index=['Car_payment','Internet_bill', 'Phone_bills', 'Home_rent','Electricity_Bill','Tax_Payment','Home_loan','food_payment']
explode=(0.2,0.2,0.1,0.05,0.2,0.3,0.3,0.2)
plt.pie(prices, labels=index, explode=explode, shadow=True, wedgeprops={'edgecolor':'black'}, radius=1)
plt.title('Random Monthly Bills')
plt.tight_layout()
plt.show()
```

```python
#third graph-> Bar-Chart
girls=[]
boys=[]
width=1
colleges=['A','B','C','D','E','F','G','H','I','J']
for i in range(10):
    boys.append(np.random.randint(1000, 3000))
    girls.append(np.random.randint(500,1000))
plt.figure(figsize=(10,6))
plt.bar(colleges , boys, color='blue', label='Boys')
plt.bar(colleges , girls, color='green', label='Girls')
plt.title('Population of Colleges')
plt.xlabel('Colleges')
plt.ylabel('Number of Students')
plt.legend()
plt.grid()
```
