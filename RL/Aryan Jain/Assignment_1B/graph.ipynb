```python
import numpy as np
import matplotlib.pyplot as plt
import time
%matplotlib inline
print("Done importing")
```

```python
x=[]
y=[]
for i in range(100,1001):
    x.append(i)
    a=np.random.rand(i,i)
    b=np.random.rand(i,i)
    t=time.time()
    m=np.dot(a,b)
    t=time.time()-t
    y.append(t*1000)
print("setting of data done")

```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel("rows in sq matrix")
ax.set_ylabel("time taken (milliseconds)")
```

```python
X=[]
Y=[]
for i in range(100,1001):
    X.append(i)
    a=np.random.rand(i,i)
    t=time.time()
    m=np.linalg.inv(a)
    t=time.time()-t
    Y.append(t*1000)
print("setting of data done")
fig,ax= plt.subplots()
ax.plot(X,Y)
ax.set_xlabel("rows in sq matrix")
ax.set_ylabel("time taken in milliseconds")
```

similar for summation of matrix


```python
X=[]
Y=[]
for i in range(100,1001):
    X.append(i)
    a=np.random.rand(i,i)
    b=np.random.rand(i,i)
    t=time.time()
    m=np.add(a,b)
    t=time.time()-t
    Y.append(t*1000)
print("setting of data done")
fig,ax= plt.subplots()
ax.plot(X,Y)
ax.set_xlabel("rows in sq matrix")
ax.set_ylabel("time taken in milliseconds")
```

As the size of matrix increases, so the number of operations required to perform the desired work...  
We say the proof of this in dot and inv functions ..
